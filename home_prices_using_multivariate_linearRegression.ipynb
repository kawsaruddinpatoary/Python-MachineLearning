```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices-2.csv')
```

```python
df
```

```python
bedroomMedian = df.bedrooms.median()
bedroomMedian
```

```python
df.bedrooms = df.bedrooms.fillna(bedroomMedian)
```

```python
df
```

```python
plt.plot(df.area, df.price)
```

```python
plt.plot(df.bedrooms, df.price)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
