# Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0, 10)
y=np.arange(11, 21)
```

```python
plt.scatter(x, y, c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph of 2D')
plt.savefig('test1.jpg')
plt.show()
```

```python
y=x*x
```

```python
plt.plot(x, y, 'go-')
plt.show()
```

## Subplots

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'go--')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'rs--')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'b+-')

plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y= np.sin(x)

plt.title('Sine wave graph')
plt.plot(x,y)
plt.show()
```
