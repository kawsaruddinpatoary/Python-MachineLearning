# Loan Approval Prediction

## Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/loan_prediction.csv")
df.head()
```

## Handling null values

```python
df.isna().sum()
```

```python
df.describe()
```

```python
limit = len(df)*0.05
cols_to_drop = df.columns[df.isna().sum() <= limit]
df.dropna(subset=cols_to_drop, inplace=True)
df.isna().sum()
```

```python
df['Self_Employed'].value_counts()
```

```python
df['Self_Employed'].fillna("No", inplace=True)
```

```python
df.isna().sum()
```

```python
df["Credit_History"].fillna(1.0, inplace=True)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.head()
```

#### Drop some unnecessary columns

```python
df.drop(["Loan_ID", "Dependents", "Self_Employed", "CoapplicantIncome", "Loan_Amount_Term"], axis=1, inplace=True)
```

```python
for col in df:
    print(f"{col}: {df[col].unique()}")
```

### Applying label encoder for categorical data

```python
from sklearn.preprocessing import LabelEncoder
# Encode categorical features
encoder = LabelEncoder()
for col in df.select_dtypes(include="object"):
    df[col] = encoder.fit_transform(df[col])
```

```python
df.head()
```

```python
df['Education'] = df['Education'].replace({0:1, 1:0})
df.head()
```

### Splitting Train test data

```python
from sklearn.model_selection import train_test_split
y = df["Loan_Status"]
x = df.drop("Loan_Status", axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape
```

### Scaling data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Machine Learning

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

### Save the model

```python
import joblib
joblib.dump(model, 'model.pkl')
print("Model saved as model.pkl")
```
