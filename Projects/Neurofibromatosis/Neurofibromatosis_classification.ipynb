## Exploratory Data Analysis

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

### Handling Null Values

```python
df.isna().sum()
```

```python
max_na = 0.05 * len(df)
max_na
```

```python
cols_to_drop = df.columns[df.isna().sum()<=max_na]
df.dropna(subset=cols_to_drop, inplace=True)
df.isna().sum()
```

```python
sns.histplot(data=df, x="Age of Mother")
```

```python
sns.histplot(data=df, x="Age of Father")
```

```python
df[["Age of Mother", "Age of Father"]].describe()
```

```python
df["Age of Mother"].fillna(np.mean(df["Age of Mother"]), inplace=True)
df["Age of Father"].fillna(np.mean(df["Age of Father"]), inplace=True)
df.isna().sum()
```

```python
df1 = df.rename(columns={"Caf‚ au lait (CLS)": "CLS"})
```

```python
df1.head()
```

### Handling Outliers

```python
df1[["Age of Mother", "Age of Father", "Age at First Diagnosis"]].describe()
```

```python
sns.histplot(data=df1, x="Age of Mother")
```

```python
sns.histplot(data=df1, x="Age of Father")
```

```python
sns.histplot(data=df1, x="Age at First Diagnosis")
```

```python
mother_std = np.std(df1["Age of Mother"])
mother_mean = np.mean(df1["Age of Mother"])
print(mother_std, mother_mean)
```

```python
df2 = df1[np.abs(df['Age of Mother'] - mother_mean) < 2*mother_std]
```

```python
sns.histplot(data=df2, x="Age of Mother")
```

```python
father_twenty_fifth = df2["Age of Father"].quantile(0.25)
father_seventy_fifth = df2["Age of Father"].quantile(0.75)
father_iqr = father_seventy_fifth - father_twenty_fifth
father_low = father_twenty_fifth - (1.5*father_iqr)
father_high = father_seventy_fifth + (1.5*father_iqr)
```

```python
df3 = df2[(df2["Age of Father"]>father_low) & (df2["Age of Father"]<father_high)]
```

```python
sns.histplot(data=df3, x="Age of Father")
```

```python
df4 = df3.drop(["Age at First Diagnosis"], axis=1)
df4.shape
```

### Preparing Train Test Data

```python
target = df4["Case Type"]
features = df4.drop(["Case Type"], axis=1)
features.head()
```

```python
target
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, target, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Finding the best model

```python
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

def find_best_classification_model_using_gridsearchcv(X, y):
    algos = {
        'logistic_regression' : {
            'model': LogisticRegression(solver='liblinear'),
            'params': {
                'C': [0.1, 1, 10],
                'penalty': ['l1', 'l2']
            }
        },
        'svc': {
            'model': SVC(),
            'params': {
                'C': [0.1, 1, 10],
                'kernel': ['linear', 'rbf']
            }
        },
        'decision_tree': {
            'model': DecisionTreeClassifier(),
            'params': {
                'criterion': ['gini', 'entropy'],
                'splitter': ['best', 'random']
            }
        },
        'random_forest': {
            'model': RandomForestClassifier(),
            'params': {
                'n_estimators': [10, 50],
                'criterion': ['gini', 'entropy']
            }
        },
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
```

```python
find_best_classification_model_using_gridsearchcv(x_scaled, target)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
